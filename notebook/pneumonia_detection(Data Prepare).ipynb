```python
import numpy as np
import pandas as pd 
import tensorflow as tf

from pathlib import Path
from matplotlib import pyplot as plt 
```

## IMPORT & EXPLORE

#### FUNCTIONS

```python
def directory_files( directory,
                     item_type = None):

    directory_files = sorted([ str(path) for path in directory.glob(f'*.{item_type}') ])

    return directory_files
```

```python
def xray_labeling( files ):

    labels = [ 0 if 'NORMAL' in path else
               1 if 'PNEUMONIA' in path 
               else -1 for path in files]
    
    return np.array(labels)
```

```python
def load_images(image_paths, target_size=(224, 224)):
    loaded_images = []
    for path in image_paths:
        try:
            # Read and decode the image as grayscale (force 1 channel)
            image = tf.io.read_file(path)
            image = tf.image.decode_image(image, channels=1)  # Force grayscale format

            # Resize the image to the target size
            image = tf.image.resize(image, target_size)

            # Normalize the pixel values to [0, 1]
            image = image / 255.0

            # Set the shape explicitly to ensure consistency
            image.set_shape((target_size[0], target_size[1], 1))

            loaded_images.append(image)
        except Exception as e:
            print(f"Error loading image at {path}: {e}")

    return loaded_images
```

```python
def display_examples(images, labels, num_examples=5):
    plt.figure(figsize=(10, 10))
    
    for i in range(num_examples):
        ax = plt.subplot(1, num_examples, i + 1)
        ax.imshow(images[i])  # Display the image
        ax.set_title(f"Label: {labels[i]}")  # Display the label
        ax.axis("off")  # Hide axes
    
    plt.show()
```

## LOAD & PREPARE

#### IMAGE DIRECTORY

```python
normal_scans_directory = Path(r"C:\Users\User\Desktop\pneumonia_detection_project\data\images\NORMAL")
pneumonia_scans_directory = Path(r"C:\Users\User\Desktop\pneumonia_detection_project\data\images\PNEUMONIA")
```

```python
normal_scans_directory_files = directory_files( directory = normal_scans_directory,
                                                item_type = 'jpeg')
pneumonia_scans_directory_files = directory_files( directory = pneumonia_scans_directory,
                                                   item_type = 'jpeg')


lung_scans_directory_files = pd.Series( (normal_scans_directory_files + pneumonia_scans_directory_files) ).sample(frac = 1,
                                                                                                                  replace = False ,
                                                                                                                  random_state = 42)
```

```python
lung_scans_directory_files = list(lung_scans_directory_files)
```

```python
lung_scans_directory_files
```

#### LABELING

```python
lung_scans_labels = xray_labeling( lung_scans_directory_files )
```

```python
lung_scans_labels
```

#### CONVERTING

```python
lung_scans = np.array(load_images( image_paths = lung_scans_directory_files ) )
lung_scans_labels = lung_scans_labels
```

#### DISPLAY

```python
display_examples(images = lung_scans, 
                 labels = lung_scans_labels, 
                 num_examples=12)
```
